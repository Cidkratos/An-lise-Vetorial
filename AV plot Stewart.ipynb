```python
import matplotlib.pyplot as plt
import numpy as np 
from sympy import *
%matplotlib inline
```

```python
x = np.linspace(0,5,10000)
```

```python
def fun(x):
    return (x-1)**4 -4*(x-1)**2
```

```python
y_graf = fun(x)
```

```python
len(y_graf)
```

```python
plt.style.use('seaborn-whitegrid')
fig, ax = plt.subplots(figsize=(10,6))
ax.plot(x,y_graf)
ax.set(xlim=(0, 5), ylim=(-5,6))
plt.xlabel('eixo X')
plt.ylabel("eixo Y")
```

```python
def x(t):
    return 1 + np.sqrt(t)
```

```python
t = np.linspace(0,5,10000)
```

```python
x_t = x(t)
x_t
```

```python
def y(t):
    return t**2 - 4*t
```

```python
y_t = y(t)
len(y_t)
```

```python
plt.style.use('seaborn-whitegrid')
fig, ax = plt.subplots(figsize=(10,6))
ax.plot(x_t,y_t,color = 'red')
ax.set(xlim=(-3, 5), ylim=(-5,6))
plt.axhline(0)
plt.axvline(0)
plt.plot(1 + np.sqrt(5),5,marker = 'o')
plt.xlabel('eixo X')
plt.ylabel("eixo Y")
```

```python
def y(x): return 2*((x+5)/3)+1
```

```python
x = np.linspace(-10,2)
```

```python
y_x = y(x)
```

```python
plt.style.use('seaborn-whitegrid')
fig, ax = plt.subplots(figsize=(10,6))
ax.plot(x,y_x,color = 'red')
ax.set(xlim=(-10, 5), ylim=(-5,6))
plt.axhline(0)
plt.axvline(0)
plt.plot(-2,3,marker = 'v',color = 'black')
plt.plot(1,5,marker = 'v',color = 'black')
plt.plot(-5,1,marker = 'v',color = 'black')
plt.plot(-8,-1,marker = 'v',color = 'black')
plt.xlabel('eixo X')
plt.ylabel("eixo Y")
```

```python
def x(y):
    return ((5-y)/2)**2 - 2
```

```python
y = np.linspace(-3,11)
```

```python
plt.style.use('seaborn-whitegrid')
fig, ax = plt.subplots(figsize=(10,6))
ax.plot(x(y),y,color = 'red')
ax.set(xlim=(-3, 15), ylim=(-4,12))
plt.axhline(0)
plt.axvline(0)
plt.plot(7,11,marker = 'o',color = 'black')
plt.plot(-2,5,marker = 'o',color = 'black')
plt.plot(4.25,0,marker = 'o',color = 'black')
plt.plot(14,-3,marker = 'o',color = 'black')
plt.plot(10,-1.9,marker = '>',color = 'black')
plt.xlabel('eixo X')
plt.ylabel("eixo Y")
```
